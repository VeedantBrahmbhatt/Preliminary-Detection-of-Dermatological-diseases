```python
#Preprocessing and data manip...
import pandas as pd
import numpy as np
from PIL import Image
import numpy as np
from sklearn.model_selection import train_test_split
## Code for mentioning file path for labels.csv
labels_df = pd.read_csv('result.csv')
class_labels = labels_df.columns[1:]
image_filenames = labels_df['image']
for i in range(len(image_filenames)):
    image_filenames[i]=image_filenames[i]+".jpg"
# print(image_filenames[0])
labels = labels_df[class_labels].values
X_train, X_val, y_train, y_val = train_test_split(image_filenames, labels, test_size=0.2, random_state=42)
X_train_df=X_train.copy()
X_train_df=pd.DataFrame(X_train_df)
X_train_df.head()
def preprocess_image(image_path, target_size=(224, 224)):
    img = Image.open('/Users/veedantbrahmbhatt/Pycharm/py/SIH/extracted_images/ISIC_2019_Training_Input/'+image_path)
    img = img.resize(target_size)
    img_array = np.array(img) / 255.0
    return img_array
X_train_processed=[]
X_val_processed=[]
for train_image_path in X_train:
    x=preprocess_image(train_image_path)
    X_train_processed.append(x)

for val_image_path in X_val:
    y=preprocess_image(val_image_path)
    X_val_processed.append(y)


X_train_processed_df=np.array(X_train_processed)
X_val_processed_df=np.array(X_val_processed)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from tensorflow.keras.applications import VGG16
from keras.models import Sequential
from keras.layers import Flatten, Dense


base_model = VGG16(weights='/Users/veedantbrahmbhatt/Pycharm/py/SIH/vgg16_weights_tf_dim_ordering_tf_kernels_notop.h5', include_top=False, input_shape=(224, 224, 3))
model = Sequential()
model.add(base_model)
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(9,activation='softmax'))
base_model.trainable=False
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

epochs = 10
batch_size = 32
history = model.fit(X_train_processed_df),  # Convert input data to NumPy array
                    y_train,
                    epochs=epochs,
                    batch_size=batch_size,
                    validation_data=((X_val_processed_df), y_val)

val_loss, val_accuracy = model.evaluate((X_val_processed_df), y_val)
print(f"Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_accuracy:.4f}")

# predictions = model.predict(np.array(X_test_processed))

# predicted_labels = np.argmax(predictions, axis=1)
```
